```python
l=[1,2,3,4]
sum_of_list=sum(l)
count=len(l)
print("sum", sum_of_list) 
print("count", count)
```

```python
M=[14,42,17,19]
sum_of_M=sum(M)
count_of_M=len(M)
print("sum_of_M =", sum_of_M)
print("count_of_M =", count_of_M)
```

```python
M=[14,42,17,19,67,90,45,23]
sum_of_M=sum(M)
count_of_M=len(M)
mean=sum_of_M/count_of_M
print("sum_of_M =", sum_of_M)
print("count_of_M =", count_of_M)
print("mean=", mean)
```
